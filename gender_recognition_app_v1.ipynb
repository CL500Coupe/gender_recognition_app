```python
import numpy as np #
import pandas as pd 
from glob import glob
import os
import PIL
import tensorflow as tf
import pathlib 
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from keras.preprocessing.image import load_img
import cv2
import random
import matplotlib.pyplot as plt
```

```python
my_model = tf.keras.models.load_model('C:/Users/macie/Desktop/machine_learning/gender_recognition_app/models/model1/gender_classification_v1_03.h5')
```

```python
class_names = ['men','women']
img_height = 180 
img_width  = 180

test_photo_dir_path=('C:/Users/macie/Desktop/machine_learning/gender_recognition_app/code/test_photos')
files=os.listdir(test_photo_dir_path)  #LIST OF PHOTOS
random_photo=random.choice(files)
correct_random_photo_path = os.path.join(test_photo_dir_path,random_photo)

test_photo_path =('test_photos/12.jpg')

img = keras.preprocessing.image.load_img(
    correct_random_photo_path, target_size=(img_height, img_width))


img_array = keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) # Create a batch

predictions = my_model.predict(img_array)
score = tf.nn.softmax(predictions[0])
```

```python
print("{} with a {:.2f}% confidence."
      .format(class_names[np.argmax(score)], 100 * np.max(score)))
img
```
